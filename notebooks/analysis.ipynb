#### Companies invest heavily in marketing campaigns across multiple platforms but lack clear insights into which campaigns, channels, and strategies generate the highest return on investment (ROI).
#### This project analyzes marketing performance across Nykaa, Purplle, and Tira to optimize campaign effectiveness and budget allocation.

```python
import pandas as pd
import numpy as np
```

```python
nykaa = pd.read_csv('../data/nykaa_campaign_data.csv')
purplle = pd.read_csv('../data/purplle_campaign_data.csv')
tira = pd.read_csv('../data/tira_campaign_data.csv')
```

```python
nykaa['Platform'] = 'Nykaa'
purplle['Platform'] = 'Purplle'
tira['Platform'] = 'Tira'
```

```python
df = pd.concat([nykaa, purplle, tira], ignore_index=True)
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['CTR'] = df['Clicks'] / df['Impressions']
df['Conversion_Rate'] = df['Conversions'] / df['Clicks']
df['Cost_per_Conversion'] = df['Acquisition_Cost'] / df['Conversions']
```

```python
# Platform Comparison

df.groupby('Platform')['Revenue'].sum()
```

#### Insights
- Nykaa generates the highest total revenue (28.66 billion), followed closely by Purplle (28.54 billion) and Tira (28.45 billion).

#### Business Impact
- Prioritize budget allocation towards Nykaa platform campaigns to leverage its revenue-generating potential and optimize overall marketing ROI.

```python
# Channel Performance

df.groupby('Channel_Used')['Conversions'].sum()
```

#### Insights
- Email channels show the highest total conversions, indicating strong performance in direct communication strategies.

#### Business Impact
- Invest more in email marketing campaigns to drive higher conversion rates and customer acquisition.

```python
# Campaign Type Analysis

df.groupby('Campaign_Type')['ROI'].mean()
```

#### Insights
- Paid Ads have the highest average ROI (2.71), followed closely by Social Media (2.71) and SEO (2.69).

#### Business Impact
- Focus budget on Paid Ads campaigns to maximize return on investment across marketing efforts.

```python
# ROI Analysis

df.sort_values(by='ROI', ascending=False)
```

#### Insights
- Sorting campaigns by ROI descending reveals the top-performing campaigns in terms of return on investment.

#### Business Impact
- Prioritize and replicate strategies from high-ROI campaigns to improve overall marketing efficiency.

```python
# Engagement Impact

df[['Engagement_Score','Conversions']].corr()
```

#### Insights
- There is a positive correlation (0.64) between Engagement Score and Conversions, indicating that higher engagement leads to more conversions.

#### Business Impact
- Focus on strategies that boost engagement scores to increase conversion rates and overall campaign effectiveness.

```python
# Platform Efficiency

df.groupby('Platform')[['CTR','Conversion_Rate']].mean()
```

#### Insights
- All platforms show similar average CTR (around 0.085) and Conversion Rates (around 0.22), with Nykaa having a slight edge.

#### Business Impact
- Since platform efficiency is comparable, choose platforms based on audience reach, cost, and strategic alignment rather than performance metrics alone.

```python
# Cost Analysis

df.groupby('Channel_Used')['Cost_per_Conversion'].mean()
```

#### Insights
- Cost per conversion varies significantly across channel combinations, with some email and social media combinations showing lower costs.

#### Business Impact
- Optimize channel selection by prioritizing cost-effective combinations to reduce acquisition costs while maintaining conversion rates.
